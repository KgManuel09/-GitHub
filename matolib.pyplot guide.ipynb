```python
# Guide for graph for teapots in GitHub.

# Warning! To you need Jupyter Notebook for working with graph.

# Step 1: import matolib.pyplot as plt.

import matplotlib.pyplot as plt  

# Step 2: create a list with coordinats. Two list x and y or one list with data

x = [23, 56, 67]

y = [13, 46, 71]

# Step 3: draw the graph 

plt.plot(x, y)

plt.show()
```

```python
# Try for var data

import matplotlib.pyplot as plt

data = [23, 53, 45, 67, 43, 10]

plt.plot(data)

plt.show()
```

```python
# You can add symbols

import matplotlib.pyplot as plt

data = [23, 53, 45, 67, 43, 10]

plt.plot(data)

# add ylabel and xlabel

plt.xlabel('x axis')

plt.ylabel('y axis')

plt.show()

```

```python
# add legend

import matplotlib.pyplot as plt  

x = [23, 56, 67, 42, 87]

y = [13, 46, 71, 51, 58] 

plt.plot(x)

plt.plot(y)

plt.legend(['blue', 'orange'], loc='upper left')

plt.show()
```

```python
# use numpy

import matplotlib.pyplot as plt

import numpy as np

x = np.array([12, 34, 65, 45])

y = np.array([35, 67, 34, 57])

plt.plot(x)

plt.plot(y)

plt.show()
```
